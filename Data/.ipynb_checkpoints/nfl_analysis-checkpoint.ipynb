# NFL v datech

## Něco málo o Americkém fotbalu
Pro správné pochopení analyzovaných dat a metrik je nezbytné seznámit se se základními principy, na kterých americký fotbal funguje. Jedná se o vysoce strategický a taktický sport, který se hraje na obdélníkovém hřišti o délce 100 yardů (cca 91 metrů). Proti sobě vždy stojí 11 hráčů útočícího týmu a 11 hráčů bránícího týmu.

Cílem útočícího týmu je posouvat šišatý míč směrem k soupeřově koncové zóně (End Zone) a skórovat. Nejběžnější způsoby bodování jsou:
* **Touchdown (6 bodů):** Hráč donese nebo chytí míč přímo v koncové zóně.
* **Field Goal (3 body):** Speciální hráč (Kicker) kopne míč mezi tyče branky.

### Mechanika hry: „Pokusy“ a zisk yardů
Tato herní mechanika je pro analýzu dat (dataset *Play-by-Play*) naprosto klíčová. Hra není plynulá jako klasický fotbal nebo hokej, ale je rozdělena na jednotlivé, striktně oddělené rozehrávky (tzv. **Plays**).

Útok má vždy **4 pokusy (Downs)** na to, aby posunul míč o minimálně **10 yardů** dopředu. 
* Pokud se mu to během těchto čtyř pokusů podaří, získává „First Down“ (nový první pokus) a dalších 10 yardů k překonání.
* Pokud to nedokáže, míč automaticky získává soupeř. 

### Typy ofenzivních akcí
Míč lze v rámci jednoho pokusu posunout vpřed dvěma základními způsoby (v datech sloupec `play_type`):
1. **Přihrávka (Pass):** Klíčový hráč zvaný *Quarterback (QB)* hodí míč vzduchem svému spoluhráči (např. *Wide Receiverovi*). Jedná se o riskantnější akci, která však může přinést velký zisk yardů najednou.
2. **Běh (Run):** Quarterback předá míč spoluhráči (nejčastěji *Running Backovi*), který se s ním snaží prorazit přes soupeřovu obranu po zemi. Je to bezpečnější varianta pro zisk kratší vzdálenosti.

Právě tato fázovaná (tahová) povaha hry, kde má každá jednotlivá akce jasný začátek, konec a exaktní výsledek v podobě získaných či ztracených yardů, dělá z amerického fotbalu jeden z nejlépe statisticky analyzovatelných sportů na světě.

### Charakteristika dat
Data jsou do projektu importována z lokálních souborů ve formátu `.csv`. Pro účely této práce byl zvolen **kompletní ročník sezóny 2025**.

K analýze a uložení do databáze byly vybrány následující datasety:

* **Play-by-play (`play_by_play_2025.csv`):** Datová sada obsahující detailní záznam o každé jednotlivé rozehrávce (play) ze všech zápasů sezóny 2025 (včetně play-off). Tento dataset primárně slouží k vyhodnocování úspěšnosti týmů, získaných yardů a efektivity konverze downů. V relaci s dalšími kolekcemi umožňuje detailní sledování výkonnosti konkrétních hráčů v rámci jednotlivých akcí.
* **Roster (`roster_2025.csv`):** Kompletní soupisky (profily) hráčů všech týmů pro sezónu 2025.
* **Games (`games.csv`):** Záznamy a základní údaje o všech odehraných zápasech (kdo s kým hrál, výsledek, typ stadionu). Z tohoto historického datasetu byla datová sada vyfiltrována čistě na ročník 2025.
* **Player Stats (`stats_player_regpost_2025.csv`):** Agregované celosezónní statistiky individuálních hráčů (součet za základní část i play-off).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

pd.set_option('display.max_columns', None)

print("Loading data from folder...")

try:
    df_pbp = pd.read_csv("play_by_play_2025.csv", low_memory=False)
    df_roster = pd.read_csv("roster_2025.csv", low_memory=False)
    df_games = pd.read_csv("games.csv", low_memory=False)
    df_stats_player = pd.read_csv("stats_player_regpost_2025.csv", low_memory=False)
    df_teams = pd.read_csv("teams_colors_logos.csv", low_memory=False)
    
  
    
    print("Data successfully loaded!")
except FileNotFoundError as e:
    print(f"Error: File not found. \nStack: {e}")

print(f"Play-by-play (Plays): {df_pbp.shape[0]} rows, {df_pbp.shape[1]} columns")
print(f"Rosters (Players): {df_roster.shape[0]} rows, {df_roster.shape[1]} columns")
print(f"Games (Games): {df_games.shape[0]} rows, {df_games.shape[1]} columns")
print(f"Player stats: {df_stats_player.shape[0]} rows, {df_stats_player.shape[1]} columns")
print(f"Teams: {df_teams.shape[0]} rows, {df_teams.shape[1]} columns")
```

# Čištění dat
V této fázi provádíme přípravu a čištění dat pro následnou analýzu a bezproblémový import do MongoDB. Proces se skládá z následujících kroků:

    Osekání sloupců (Feature selection): Z původních obsáhlých datasetů vybíráme do nových tabulek pouze úzce specifikovaný seznam sloupců, které nesou informační hodnotu pro náš projekt.

    Analýza chybějících hodnot: Pomocí funkce isna().sum() zjišťujeme počty nevyplněných záznamů v jednotlivých sloupcích.

    Kontextové nahrazování: Nevyplněné hodnoty u zisku yardů (yards_gained, ydstogo) jsou nahrazeny 0. U ostatních atributů (např. identifikátory hráčů či výška) ponecháváme systémový NaN, abychom nezkreslili statistická data. Tento přístup zároveň optimalizuje databázi MongoDB, která chybějící hodnoty při importu inteligentně ignoruje.

    Odstranění "šumu": Odstraňujeme řádky představující technické nebo reklamní přestávky, které jsou běžnou součástí surových dat z NFL, ale nemají vliv na sportovní statistiky.

    Kontrola: Pomocí metody head(3) si u každého datasetu vypisujeme náhled upravených dat.

```python
# Column cleaning

# Play by play
pbp_cols = ['game_id', 'play_id', 'play_type', 'posteam', 'defteam', 'down', 'ydstogo', 
            'yards_gained', 'passer_player_id', 'receiver_player_id', 'rusher_player_id']
df_pbp_clean = df_pbp[pbp_cols].copy()

print("--- Empty values in dataset columns (Play-by-Play) ---")
print(df_pbp_clean.isna().sum())

# set 0 yard when empty yards columns
df_pbp_clean['yards_gained'] = df_pbp_clean['yards_gained'].fillna(0)
df_pbp_clean['ydstogo'] = df_pbp_clean['ydstogo'].fillna(0)

# Removing plays that doesnt have type (commercials...)
df_pbp_clean = df_pbp_clean.dropna(subset=['play_type'])

print("--- Empty values in dataset columns after cleaning (Play-by-Play) ---")
print(df_pbp_clean.isna().sum())

print("\n--- Preview of cleaned data (Play-by-Play) ---")
display(df_pbp_clean.head(3))

# Roster
roster_cols = ['gsis_id', 'team', 'position', 'full_name', 'height', 'weight', 'college', 'entry_year', 'rookie_year']
df_roster_clean = df_roster[roster_cols].copy()

print("--- Empty values in dataset columns (Roster) ---")
print(df_roster_clean.isna().sum())

print("\n--- Preview of cleaned data (Roster) ---")
display(df_roster_clean.head(3))


#Player stats
stats_cols = ['player_id', 'player_display_name', 'position', 'position_group', 'recent_team', 'games', 
              'passing_yards', 'passing_tds', 'rushing_yards', 'receiving_yards', 'fantasy_points', 'sacks_suffered', 'sack_fumbles', 'sack_fumbles_lost', 'passing_first_downs']
df_stats_player_clean = df_stats_player[stats_cols].copy()

print("--- Empty values in dataset columns (Player stats) ---")
print(df_stats_player_clean.isna().sum())

print("\n--- Preview of cleaned data (Player stats) ---")
display(df_stats_player_clean.head(3))

# Games
games_cols = ['game_id', 'season', 'week', 'game_type', 'home_team', 
              'away_team', 'home_score', 'away_score', 'roof']
  # Get games only in 2025 season

df_games = df_games[df_games['season'] == 2025]
df_games_clean = df_games[games_cols].copy()

print("--- Empty values in dataset columns (Games) ---")
print(df_games_clean.isna().sum())

print("\n--- Preview of cleaned data (Games) ---")
display(df_games_clean.head(3))

#Teams
teams_cols = [
    'team_abbr',
    'team_name',
    'team_id',
    'team_conf',
    'team_division'
]

df_teams_clean = df_teams[teams_cols].copy()

print("--- Empty values in dataset columns (Teams) ---")
print(df_teams_clean.isna().sum())

print("\n--- Preview of cleaned data (Teams) ---")
display(df_teams_clean.head(3))
```

# Statistiky

V této sekci jsou vybrány ukázky základních statistických údajů při práci s datasety

### Statistika typů akcí s jejich celkovým počtem, počtem získaných yardů, průměrných ziskem za akci a nejdelší akcí
Dochází zde k vyfiltrování ofenzivních akcí (run = běh, pass = hod) pomocí hodnot sloupce "play_type".
Na vybrané sloupce se aplikuje groupby a agregační funkce (count, sum, mean, max).
Aplikování "columns" s polem stringů dojde k přepsání výchozích názvů sloupců, což přispěje k vyšší čitelnosti.
Funkce display, které předáme zaokrohuhlené statistiky seřazené sestupně podle celkového počtu akcí, zobrazí výslednou tabulku



```python
offensive_plays = ["run", "pass"]
df_pbp_offensive = df_pbp_clean[df_pbp_clean['play_type'].isin(offensive_plays)]

action_statistics = df_pbp_offensive.groupby('play_type')['yards_gained'].agg(['count', 'sum', 'mean', 'max'])
action_statistics.columns = ['Počet akcí', 'Celkem yardů', 'Průměr na akci', 'Nejdelší akce']
display(action_statistics.round(2).sort_values('Počet akcí', ascending=False))
```

#### Závěr
Během sezóny 2025 převažovaly akce typu  "pass", které vedly k většímu celkvému i průměrnému zisku yardů. Nicméně akce typu "run" zaznamenaly nejvyšší honotu zisku yardů během jedné akce.

### Top 5 Rozehrávačů za rok 2025
Při vytváření této statistiky dochází k vyfiltrování hráčů, kteří mají vůbec nějaké naházené yardy a následně se tyto hodnoty seřadí podle "passing_yards"
Poté dojde k výpočtu sloupce, který se v datasetu nenachází o to je počet touchdownů na 100 yardů.
Z důvodu čitelnosti se  přejmenují názvy sloupců funkcí "rename" a namapování klíčů k labels.
Funkce display následně vypíše tabulku se statistikami.


```python
qbs_only = df_stats_player_clean[df_stats_player_clean['passing_yards'] > 0]

top_passers = qbs_only.sort_values(by='passing_yards', ascending=False).head(5).copy()

top_passers['tds_per_100'] = (top_passers['passing_tds'] / top_passers['passing_yards']) * 100

top_passers['tds_per_100'] = top_passers['tds_per_100'].round(2)

top_passers = top_passers.rename(columns={
    'player_display_name': 'Jméno', 
    'recent_team': 'Tým', 
    'passing_yards': 'Passing yards', 
    'passing_tds': 'Passing TDs',
    'tds_per_100': 'TDs per 100 yds'
})

print("\n--- TOP 5 QBs (Passing yards 2025) ---")
display(top_passers[['Jméno', 'Tým', 'Passing yards', 'Passing TDs', 'TDs per 100 yds']])
```

#### Závěr
Rozehrávačem s nejvíce naházenými yardy byl Matthew Stafford z týmu LA Rams (5643 yardů), který zaznamenal i nejvíce "passing" touchdownů (52). Mezi prvním a druhým místem byl skok v počtu naházených touchdownů. Z výsledků statistiky můžeme dále pozorovat, že počet naházených yardů není přímo úměrný počtu naházených touchdownů. Například u Jareda Goffa, který má a přibližně 700 naházených yardů míň, je vidět vyšší počet touchdownů na 100 xardů než u druhého Drake Mayeho.

### Průměrný počet týmu za hru Domácí vs Hostující tým
V této statistice dochází k ukázání poměru průměrného zisku bodů za hru mezi domácím a hostujícím týmem.
Z čistého games datasetu se vyberou sloupce reprezentující skóre domácích týmů a hostujících týmu a z nich se vypočítá zaokrouhlený průměr.


```python
avg_scores = df_games_clean[['home_score', 'away_score']].mean().round(2)
avg_scores.index = ['Domácí (Home)', 'Hosté (Away)']
display(avg_scores)
```

#### Závěr
Z výsledné statistiky můžeme pozorovat, že domácí týmy mají lehce větší tendenci skórovat v průměru více bodů než týmy hostující.

### 10 Nejhrannějších pozic v NFL (největší počet hráčů na soupisce týmů na dané pozici)
Tato statistika má reprezentovat ukázku pozic s největším zastoupením hráčů na soupiskách jednotlivých týmů

```python
pos_counts = df_roster_clean['position'].value_counts().head(10)
display(pos_counts)
```

#### Závěr
Z výsledků statistiky můžeme vidět, že pozice s největším zasoupením hráčů na soupiských je DB neboli Defensive back, Druhá v pořadí je offensive lineman.
Na posledním 10. místě naší statistiky se nachází pozice LS (long snapper), kterých je diametrálně odlišně oproti DB.

# Grafická vizualizace
Pro lepší vhled do struktury dat a objevení základních závislostí byly vytvořeny dvě vizualizace reprezentující klíčové herní mechanismy.

* **Graf 1: Rozložení ofenzivních akcí (Pass vs. Run)**
  Koláčový graf znázorňuje globální balanc útočné fáze hry v sezóně 2025. Vizuálně potvrzuje poměr, v jakém se v moderní NFL využívá riskantnější hra vzduchem (přihrávky) oproti bezpečnější hře po zemi (běhy).
  * Z grafu můžeme získat následující informace:
    * **Dominance vzdušné hry:** Přestože se to může sezónu od sezóny mírně lišit, graf potvrzuje, že moderní NFL je tzv. *„pass-heavy“* liga (orientovaná na přihrávky). Přihrávky standardně tvoří nadpoloviční většinu ofenzivních akcí (často nad 55 %).
    * **Strategie týmů:** Důvodem pro tento poměr je fakt, že úspěšná přihrávka statisticky generuje více yardů na jednu akci než běh. Běhová hra (Run) je naopak využívána ke kontrole času, bezpečnějšímu zisku krátkých vzdáleností a k „unavení“ obrany.



**Graf 2: Korelace výkonu Quarterbacků (Bodový graf)**
  Tento bodový graf (*scatter plot*) analyzuje vztah mezi celkovým počtem naházených yardů a počtem dosažených touchdownů u jednotlivých quarterbacků. Z vyobrazených dat lze jasně pozorovat silnou korelaci – u hráčů je patrný lineární trend, kdy vyšší objem získaných yardů spolehlivě predikuje i vyšší počet skórujících akcí.
* Z grafu můžeme získat následující informace:
  * **Silná přímá úměra (Korelace):** Z bodového grafu je jasně patrný stoupající lineární trend. Čím více yardů quarterback nahází (pohyb po ose X doprava), tím více touchdownů vyprodukuje (pohyb po ose Y nahoru).
    * **Identifikace anomálií (Outliers):** Graf nám umožňuje snadno najít výjimky z pravidla:
      * *Vysoká efektivita (Bod vysoko, ale spíše vlevo):* Quarterback sice nenaházel celkově tolik yardů, ale když už se tým blížil ke koncové zóně (Red Zone), dokázal akci úspěšně zakončit touchdownem.
      * *Problém se zakončením (Bod hluboko vpravo, ale nízko):* Quarterback sice dokáže posouvat tým po hřišti (vysoký počet yardů), ale v klíčových momentech u koncové zóny tým selhává (např. útok končí pouze Field Goaly za 3 body místo Touchdownů za 6 bodů, nebo tým přenechává skórování běžcům).
    * **Rozdělení hráčů:** Shluk bodů v levém dolním rohu představuje náhradníky (nebo zraněné hráče), kteří odehráli jen zlomek zápasů. Naopak izolované body v pravém horním rohu patří absolutní elitě ligy pro danou sezónu.

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
play_counts = df_pbp_offensive['play_type'].value_counts()
plt.pie(play_counts, labels=play_counts.index.str.upper(), autopct='%1.1f%%', 
        colors=['#66b3ff', '#99ff99'], startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Pass vs Run', fontsize=14)

# GRAF 2: Bodový graf - Naházené yardy vs Touchdowny u Quarterbacků (z tabulky Player Stats)
plt.subplot(1, 2, 2)
# Vyfiltrujeme jen hráče s pozicí QB, kteří hodili alespoň 1 yard
qbs = df_stats_player_clean[(df_stats_player_clean['position'] == 'QB') & (df_stats_player_clean['passing_yards'] > 0)]
plt.scatter(qbs['passing_yards'], qbs['passing_tds'], color='coral', alpha=0.7, edgecolors='black')
plt.title('Quarterbacks: Passing Yards vs Passing TDs', fontsize=14)
plt.xlabel('Naházené yardy (Passing Yards)')
plt.ylabel('Počet Touchdownů (Passing TDs)')
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
